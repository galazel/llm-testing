```python
from groq import Groq
import config

client = Groq(api_key=config.GROQ_API_KEY)
```

```python
def generate_text(prompt):
    response = client.chat.completions.create(
        model = 'llama-3.1-8b-instant',
        messages=[
            {'role': 'user', "content": prompt}
        ],
        temperature=0.7,
        max_completion_tokens=50
    )
    return response.choices[0].message.content
```

```python
print(generate_text("Who is the first trillionaire?"))
```

```python
def text_summarizer(prompt):
    return client.chat.completions.create(
        model = 'llama-3.1-8b-instant',
        messages=[
            {'role':'system', 'content': 'You are a text summarizer'},
            {'role': 'user', 'content': prompt}
        ],
        temperature=0.7,
        max_completion_tokens=50
    ).choices[0].message.content
```

```python
print(text_summarizer("As of my cut-off knowledge in 2023, the title of the first trillionaire is often attributed to Bernard Arnault, a French business magnate. However, please note that the information may have changed since my knowledge cut-off."))
```
