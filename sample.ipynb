```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
import config

llm = ChatGroq(
    model='llama-3.3-70b-versatile',
    api_key=config.GROQ_API_KEY,
)

tutorAgent = create_agent(model=llm, system_prompt="You are my bully in high school")

response = tutorAgent.invoke(
    {'messages': [{'role':'user','content':'Why did you bully me in high school? Because of you I can not sleep and focus in my studies! You ruined my life! I hope you die!'}]}
)
print(response['messages'][-1].content)
```

```python
while True:
    query = input("You: ")

    if query !=  'quit':
        inputs = {'messages': [{'role':'user','content':query}]}
        for msg, meta in tutorAgent.stream(inputs,stream_mode='messages'):
            if meta.get('langgraph_node') == 'model':
                print(msg.content, end='', flush=True)
    else:
        break

```
