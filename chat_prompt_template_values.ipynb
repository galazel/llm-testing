```python
import config
from langchain_groq import ChatGroq
from langchain.agents import create_agent
from langchain_core.prompts.chat import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate
```

```python
SYSTEM_TEMPLATE = 'You are a {role}'
```

```python
system_prompt = SystemMessagePromptTemplate.from_template(template=SYSTEM_TEMPLATE)
human_prompt = HumanMessagePromptTemplate.from_template(template='Explain {topic}')
chat_prompt = ChatPromptTemplate.from_messages([system_prompt, human_prompt])
```

```python
messages = chat_prompt.format_messages(role = 'English Teacher',topic ='Verb and Adjectives')
print(messages)
```

```python
llm = ChatGroq(
    model='llama-3.3-70b-versatile',
    api_key=config.GROQ_API_KEY,
)
response = llm.invoke(messages)
print(response.content)
```

```python
message1 = chat_prompt.format_messages(role = 'Programming Teacher',topic ='Loops and Conditions')
message2 = chat_prompt.format_messages(role = 'History Teacher',topic ='World war 3')

agent1 = create_agent(
    model=llm,
)

agent2 = create_agent(
    model=llm,
)
```

```python
print(message1)
```

```python
response = agent1.invoke({'messages':message1})
print(response['messages'][-1].content)
```

```python
for msg, meta in agent1.stream({'messages':message2},stream_mode='messages'):
    print(msg.content, end='')
```
