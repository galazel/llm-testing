```python
import config
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
list_instructions = CommaSeparatedListOutputParser().get_format_instructions()
```

```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", list_instructions),
    ("user", "I have recently adopted a {pet}. Could you suggest three {pet} names?")
])
```

```python
chat_template.messages[0].prompt.template
```

```python
chat = ChatGroq(
    model='llama-3.3-70b-versatile',
    api_key=config.GROQ_API_KEY,
)
```

```python
list_output_parser = CommaSeparatedListOutputParser()
```

```python
chat_template_result = chat_template.invoke({'pet' : 'dog'})
```

```python
result = chat.invoke(chat_template_result)
```

```python
list_output_parser.invoke(result)
```

```python
chain = chat_template | chat | list_output_parser
```

```python
chain.invoke({'pet':'Dog'})
```
