```python
import config
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
```

```python
chat_template = ChatPromptTemplate.from_messages([
    ('human','I had studied {subject}. can you give me roadmap for that or tips')
])
```

```python
chat = ChatGroq(
    model='llama-3.3-70b-versatile',
    api_key=config.GROQ_API_KEY,
    max_tokens=100
)
```

```python
chain = chat_template | chat
```

```python
chain.invoke({'subject':'Springboot'})
```

```python
chain.batch([{'subject':'springboot'},{'subject':'mysql'}])
```

```python
for i in chain.stream({'subject':'springboot'}):
    print(i.content, end='', flush=True)
```
