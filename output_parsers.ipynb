```python
import config
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import StrOutputParser, CommaSeparatedListOutputParser
```

```python
llm = ChatGroq(
    model='llama-3.3-70b-versatile',
    api_key=config.GROQ_API_KEY,
    temperature=0.70
)
```

```python
message_h = HumanMessage(content='Give me interesting facts')
```

```python
response = llm.invoke([message_h])
```

```python
str_output_parser = StrOutputParser()
```

```python
response_parsed = str_output_parser.invoke(response)
print(response_parsed)
```

```python
message_h = HumanMessage(content=f'I have a dog but I do not know what name to be given. Can you give me some. {CommaSeparatedListOutputParser().get_format_instructions()}')
response = llm.invoke([message_h])

comma_parser = CommaSeparatedListOutputParser()
comma_parsed = comma_parser.invoke(response)
print(comma_parsed)
```
