RunnablePassthrough

```python
import config
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableParallel, chain
```

```python
RunnablePassthrough().invoke('hi')
```

```python
chat_template = ChatPromptTemplate.from_messages([
    ('human','List the most essential tools for {profession}. Answer only by listing the tools')
])
```

```python
chat_template_strategy = ChatPromptTemplate.from_messages([
    ('human','Considering the tools provided, develop a strategy for effectively learning and '
             'mastering them')
])
```

```python
chat = ChatGroq(
    model='llama-3.3-70b-versatile',
    api_key=config.GROQ_API_KEY,
    max_tokens=100
)
```

```python
str_parser = StrOutputParser()
```

```python
chain_s = chat_template | chat | str_parser | {'tools':RunnablePassthrough()} | chat_template_strategy | chat | str_parser
```

```python
result = chain_s.invoke({"profession":"Frontend Developer"})
```

RunnableParallel

```python
chat_template_books = ChatPromptTemplate.from_messages([ ('human','Suggest three of the best intermediate-level {programming language} books. Answer only by listing the books.')
])
chat_template_projects = ChatPromptTemplate.from_messages([
    ('human','Suggest three interesting {programming language} projects suitable for intermediate-level programmers.')
])
```

```python
@chain
def to_upper(x):
    return x.upper()
```

```python
chain_books  = chat_template_books | chat | str_parser
chain_projects = chat_template_projects | chat | str_parser | to_upper
```

```python
chain_parallel = RunnableParallel({'books':chain_books,'projects':chain_projects})
```

```python
chain_parallel.invoke({'programming language':'JavaScript'})
```
