```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters.character import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
loader = PyPDFLoader('Galagar_Glyzel_Profis10_Midterm_Project.pdf')
```

```python
document = loader.load()

for page in document:
    page.page_content = ' '.join(page.page_content.split())
```

```python
document_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
```

```python
pages_document_split = document_splitter.split_documents(document)
```

```python
pages_document_split
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-mpnet-base-v2",
    encode_kwargs={"normalize_embeddings": True},
)
```

```python
for document in pages_document_split:
    print(embeddings.embed_query(document.page_content))
```
